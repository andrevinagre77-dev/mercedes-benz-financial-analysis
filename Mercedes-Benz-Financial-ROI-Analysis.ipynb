```python

# Technical Execution Metrics
min_price = df['Low'].min()
max_price = df['High'].max()
profit_per_share = max_price - min_price

# Identifying specific execution dates
date_min = df['Date'][df['Low'].idxmin()]
date_max = df['Date'][df['High'].idxmax()]

# Professional Execution Output
print(f"--- INVESTMENT EXECUTION SUMMARY ---")
print(f"Entry Price: €{round(min_price, 2)} | Date: {date_min}")
print(f"Exit Price:  €{round(max_price, 2)} | Date: {date_max}")
print(f"Profit Delta: €{round(profit_per_share, 2)} per share")
print(f"------------------------------------")
```

```python
import matplotlib.pyplot as plt

# Dados
labels = ['Preço de Compra', 'Preço de Venda']
valores = [17.48, 77.90]
lucro = 60.42

fig, ax = plt.subplots(figsize=(8, 6))

# Criar as barras
bars = ax.bar(labels, valores, color=['#ff9999', '#66b3ff'])

# Adicionar a linha de lucro entre as duas barras
plt.annotate('', xy=(1, 77.90), xytext=(1, 17.48),
             arrowprops=dict(arrowstyle='<->', color='green', lw=2))
plt.text(1.05, 47, f'Lucro: +{lucro}€\n(+345%)', color='green', fontweight='bold', fontsize=12)

# Estilização
plt.title('Performance da Ação: 2020 vs 2022', fontsize=14)
plt.ylabel('Valor em Euros (€)')
ax.set_ylim(0, 90)

# Adicionar os valores exatos por cima das barras
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 2, f'{yval}€', ha='center', fontweight='bold')

plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Investment Operation Data
initial_investment = 1000
net_profit = 3456.50  # Based on ~345% profit margin
final_value = initial_investment + net_profit

# Visual Configuration
plt.figure(figsize=(10, 6))
labels = ['Initial Investment', 'NET PROFIT', 'FINAL PORTFOLIO VALUE']
values = [initial_investment, net_profit, final_value]
colors = ['#546e7a', '#2e7d32', '#1565c0']  # Slate Grey, Profit Green, Portfolio Blue

# Create Bar Chart
bars = plt.bar(labels, values, color=colors, width=0.6)

# Add data labels on top of each bar for immediate clarity
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 100, 
             f'€{yval:,.2f}', ha='center', va='bottom', 
             fontweight='bold', fontsize=12)

# Professional Styling & Titles
plt.title('Investment Performance: Mercedes-Benz (MBG.DE)', fontsize=15, fontweight='bold', pad=30)
plt.ylabel('Amount in Euros (€)', fontsize=11)

# Remove chart junk (spines) for a cleaner look
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

# Add subtle grid for scale reference
plt.grid(axis='y', linestyle='--', alpha=0.3)

plt.tight_layout()
plt.show()
```
